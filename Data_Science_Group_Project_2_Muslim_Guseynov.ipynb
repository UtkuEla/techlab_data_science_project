Average Annual Income by Professional Status and Gender


```python
# Calculate the average annual income for each combination of Professional Status and Gender
job_gender_salary = df.groupby(['Professional Status', 'Gender'])['Annual Income'].mean().reset_index()

# Find the maximum value in the data for setting y-axis limits
max_value = job_gender_salary['Annual Income'].max()
print("Maximum Annual Income:", max_value)

# Plotting
plt.figure(figsize=(10, 6))
sns.barplot(x='Professional Status', y='Annual Income', hue='Gender', data=job_gender_salary, palette=['blue', 'red', 'purple'], width=0.5)
sns.set_style('darkgrid')

# Setting y-axis limits to ensure all values are visible
plt.ylim(0, max_value + 50000)

# Customizing the plot
plt.title('Average Annual Income by Professional Status and Gender')
plt.ylabel('Average Annual Income')
plt.xlabel('Professional Status')
plt.legend(title='Gender')
plt.xticks(rotation=45) 

# Show the plot
plt.show()
```

Average Frequency of Visit by Professional Status and Gender

```python
# Calculate the average annual income for each combination of Professional Status and Gender
visit_gender_salary = df.groupby(['Professional Status', 'Gender'])['Frequency of Visit'].mean().reset_index()

# Find the maximum value in the data for setting y-axis limits
max_value = visit_gender_salary['Frequency of Visit'].max()
print("Maximum Frequeny of Visit:", max_value)

# Plotting
plt.figure(figsize=(10, 6))
sns.barplot(x='Professional Status', y='Frequency of Visit', hue='Gender', data=visit_gender_salary, palette=['blue', 'red', 'purple'], width=0.5)
sns.set_style('darkgrid')

# Setting y-axis limits to ensure all values are visible
plt.ylim(0, max_value + 3)

# Customizing the plot
plt.title('Average Frequency of Visit by Professional Status and Gender')
plt.ylabel('Average Frequnecy of Visit')
plt.xlabel('Professional Status')
plt.legend(title='Gender')
plt.xticks(rotation=45) 

# Show the plot
plt.show()
```
