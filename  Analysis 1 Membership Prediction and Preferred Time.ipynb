```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r"C:\Users\iusti\Downloads\dataset_gym_membership_group2.csv")
```

```python
df
```

```python
#Preferred time by gender

pref_time=df.groupby(["Preferred_Time"])
pref_time_men=pref_time["Gender"].apply(lambda x: x.str.contains("Male").sum())
pref_time=df.groupby(["Preferred_Time"])
pref_time_fem=pref_time["Gender"].apply(lambda x: x.str.contains("Female").sum())
combined_counts = pd.DataFrame({
    'Male': pref_time_men,
    'Female': pref_time_fem
})
combined_counts_reset = combined_counts.reset_index()
combined_counts_reset
combined_counts_melted = pd.melt(combined_counts_reset, id_vars=['Preferred_Time'], value_vars=['Male', 'Female'], var_name='Gender', value_name='Count')
sns.barplot(x='Preferred_Time', y='Count', hue='Gender', data=combined_counts_melted)
plt.title('Preferred Time by Gender')
plt.show()
```

```python
#Preferred time by membership type

pref_time=df.groupby(["Preferred_Time"])
pref_time_basic=pref_time["Membership_Type"].apply(lambda x: x.str.contains("Basic").sum())
pref_time=df.groupby(["Preferred_Time"])
pref_time_premium=pref_time["Membership_Type"].apply(lambda x: x.str.contains("Premium").sum())
pref_time=df.groupby(["Preferred_Time"])
pref_time_VIP=pref_time["Membership_Type"].apply(lambda x: x.str.contains("VIP").sum())
combined_counts2 = pd.DataFrame({
    'Basic': pref_time_basic,
    'Premium': pref_time_premium,
    'VIP': pref_time_VIP
})
combined_counts2_reset = combined_counts2.reset_index()
combined_counts2_reset
combined_counts2_melted = pd.melt(combined_counts2_reset, id_vars=['Preferred_Time'], 
                                 value_vars=['Basic', 'Premium', 'VIP'],
                                 var_name='Membership_Type', value_name='Count')
sns.barplot(x='Preferred_Time', y='Count', hue='Membership_Type', data=combined_counts2_melted)
plt.title('Preferred Time by Membership Type')
plt.show()
```
